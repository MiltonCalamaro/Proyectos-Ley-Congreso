```
import pandas as pd
```

```
import os
os.chdir("/content/drive/My Drive/Colab Notebooks/Proyectos_Ley_Congreso/notebooks")
```

```
dataPdf=pd.DataFrame()
for i in os.listdir("../data/intermediated/pdf_recognized"):
  aux=pd.read_pickle("../data/intermediated/pdf_recognized/"+i)
  dataPdf=pd.concat([dataPdf,aux], axis=0)
print(dataPdf.shape)
dataPdf.head()
```

```
aux1=dataPdf.groupby(["Archivo_Origen"]).Fecha.apply(lambda x : x.tolist()[0]).to_frame()
aux2=dataPdf.groupby(["Archivo_Origen"]).Pagina.count().to_frame()
aux3=dataPdf.groupby(["Archivo_Origen"]).Texto.apply(lambda x : "\n".join(x)).to_frame()
dataPdf_=aux1.join(aux2, on="Archivo_Origen").join(aux3, on="Archivo_Origen").reset_index().rename(columns={"Pagina":"NroPaginas"})
```

```
index=dataPdf_["Archivo_Origen"].apply(lambda x: x[3:7])
dataPdf_.set_index(index,inplace=True)
dataPdf_.index.names = ['index']
print(dataPdf_.shape)
dataPdf_.head()
```

```
dataMetadata=pd.read_pickle("../data/raw/dataMetadata.pkl")
print(dataMetadata.shape)
dataMetadata.head()
```

```
dataIntegrated=pd.merge(dataMetadata,dataPdf_,how="left",on="index")
print(dataIntegrated.shape)
dataIntegrated.head()
```

```
dataIntegrated[dataIntegrated["Texto"].astype(str)!="nan"].to_pickle("../data/intermediated/dataIntegrated.pkl")
```
