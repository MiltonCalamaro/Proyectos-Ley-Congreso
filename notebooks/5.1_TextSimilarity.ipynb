```
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import os
os.chdir("/content/drive/My Drive/Colab Notebooks/Proyectos_Ley_Congreso/notebooks")
```

```
df_preprocessed=pd.read_pickle("../data/preprocessed/df_preprocessed_w2v.pkl")
```

```
promVect_train=[np.array(i) for i in df_preprocessed["word2vec"]]
```

```
### comparison with distance coseno
most_relateds=[]
for w2v in promVect_train:
  vals = cosine_similarity(w2v.reshape(1,-1), promVect_train)
  idx=vals.argsort()[0][-6:-1]
  most_related=df_preprocessed.iloc[idx].index
  
  flat = vals.flatten()
  flat.sort()
  confidence=flat[-6:-1]

  dict_={}
  for i,j in zip(most_related,confidence):
    dict_[i]=j  

  most_relateds.append(dict_)

df_preprocessed["most_relateds"]=most_relateds
df_preprocessed
```

```
df_preprocessed.to_pickle("../data/preprocessed/df_text_similarity.pkl")
```
